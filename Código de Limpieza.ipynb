```python
# Bibliotecas a importar

import pandas as pd
import numpy as np
```

```python
# Importar archivo

eng = pd.read_csv("Ingles DS3_2014_01.csv",sep=';',encoding='latin-1')
```

```python
for columna in eng.columns:
    print('Nulos en ', columna, '=', eng[columna].isna().sum())

#Data no tiene nulos en las columnas de interes, así que se hace "drop" a columnas que no se van a usar
```

```python
eng_w = eng.drop(columns=['Region_domicilio_registrado', 'localidad'])
eng_w.columns
```

```python
eng_w.dtypes
```

```python
# Se cambian los valores string de las variables 'Toeic_Inicial' y 'Toeic_Final' a 0

eng_w = eng_w.replace("No asistió", 0)
eng_w = eng_w.replace("S/I", 0)
eng_w.tail()
```

```python
# Eliminar rows que contengan variables 0, ya que no sirven para el contraste que se busca hacer

eng_w = eng_w.query('0 != Toeic_Inicial')
eng_w = eng_w.query('0 != Toeic_Final')

eng_w
```

```python
eng_w.groupby('tipo_beca').count()
```

```python
# Se cambian los valores de 'tipo_beca' para que sean todos el mismo

eng_w = eng_w.replace("100 hrs.", "100 Horas")
eng_w = eng_w.replace("200 hrs.", "200 Horas")
eng_w.groupby('tipo_beca').count()
```

```python
# Se cambian los tipos de las columnas anteriores a float, y a la columna 'tipo_beca' a string

eng_w[['Toeic_Inicial', 'Toeic_Final']] = eng_w[['Toeic_Inicial', 'Toeic_Final']].apply(pd.to_numeric)

#eng_w['tipo_beca'] = eng_w['tipo_beca'].astype(str)

eng_w.dtypes
```

```python

regiones = regiones = {1: 'Tarapaca', 2: 'Antofagasta', 3: 'Atacama', 4: 'Coquimbo', 5: 'Valparaiso', 6: "O'Higgins", 7: 'Maule', 8: 'Bio-Bio', 9: 'Araucania', 10: 'Los Lagos', 11: 'Aisen', 12: 'Magallanes', 13: 'Metropolitana', 14: 'Los Rios', 15: 'Arica y Parinacota'}

eng_w['region_nombre'] = eng['region_curso'].map(regiones)

engorg = eng_w.iloc[:, [0, 1, 5, 2, 3, 4]]  # Orden de Columnas
engorg
```

```python
# Se exporta la data trabajada

engorg.to_csv('data_powerbi_toeic.csv', encoding='utf-8', index=False, header=True)

```
